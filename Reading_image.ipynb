```python
import cv2
import matplotlib.pyplot as plt

# Path to your image in Google Drive
img_path = '/content/drive/MyDrive/Colab Notebooks/Data/PHOTO-2025-11-06-18-39-47.jpg'


# Read image
img = cv2.imread(img_path)

# Convert BGR -> RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Show image
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```
